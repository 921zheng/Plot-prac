```python
import numpy as np
```

```python
#create data
x=np.linspace(0,10,100)
x[:10]
```

```python
#plot
fig,ax=plt.subplots()
ax.plot(x,x**2)
```

```python
#scatter
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x))
```

```python
#make a plot from a dictionary
butter_price={"Almond Butter":10,
              "CashewnutButter":8,
              "Peanut Butter":6}
fig,ax=plt.subplots()
ax.bar(butter_price.keys(),butter_price.values())
```

```python
#change the direction of the bar
fig,ax=plt.subplots()
ax.barh(list(butter_price.keys()),list(butter_price.values()))
```

```python
#make some data for histogroms and plot it
x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x)
```

```python
#several plots together
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(5,10))
ax1.plot(x,x/2)
ax2.scatter(np.random.randn(10),np.random.randn(10))
ax3.bar(butter_price.keys(),butter_price.values())
ax4.hist(np.random.randn(1000));
```

```python
#different way
fig,ax=plt.subplots(ncols=2,nrows=2,figsize=(5,10))
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.randn(10),np.random.randn(10))
ax[1,0].bar(butter_price.keys(),butter_price.values())
ax[1,1].hist(np.random.randn(1000));
```
